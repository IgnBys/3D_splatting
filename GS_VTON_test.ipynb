## Data Preparation

Based on the project's requirements, you'll need to prepare your data in two main steps, using external tools, and then organize it into a specific folder structure.

### Step 1: Process Video Data with Nerfstudio

Follow the official Nerfstudio documentation to process your own video data. This typically involves converting your video into a format suitable for NeRF training. You can find detailed instructions here: [Nerfstudio Quickstart - Custom Dataset](https://docs.nerf.studio/quickstart/custom_dataset.html)

### Step 2: Obtain Initialized 3D Gaussian Splatting (3DGS)

After processing your video with Nerfstudio, you will need to obtain the initialized 3DGS. Refer to the 3D Gaussian Splatting repository for guidance on this step: [3D Gaussian Splatting GitHub](https://github.com/graphdeco-inria/gaussian-splatting)

