```python
import os

# 1. Clean the requirements file
%cd /content/GS-VTON
if os.path.exists('requirements.txt'):
    with open('requirements.txt', 'r') as f:
        lines = f.readlines()

    cuda_exts = ["tiny-cuda-nn", "nvdiffrast", "diff-gaussian-rasterization", "simple-knn"]
    already_installed = ['torch', 'torchvision', 'torchaudio', 'xformers', 'detectron2', 'basicsr']

    with open('requirements_clean.txt', 'w') as f:
        for line in lines:
            if not any(x in line.lower() for x in cuda_exts + already_installed):
                f.write(line)

# 2. Force a clean rebuild of tiny-cuda-nn and nvdiffrast
# Using --force-reinstall and --no-cache-dir to fix the 'undefined symbol' error
print("Installing nvdiffrast...")
!pip install --no-build-isolation --force-reinstall --no-cache-dir git+https://github.com/NVlabs/nvdiffrast.git

print("Installing tiny-cuda-nn (Clean Rebuild - takes ~5 mins)...")
!pip install --no-build-isolation --force-reinstall --no-cache-dir git+https://github.com/NVlabs/tiny-cuda-nn/#subdirectory=bindings/torch

# 3. Install the remaining standard dependencies
print("Installing remaining dependencies...")
!pip install -r requirements_clean.txt
```
